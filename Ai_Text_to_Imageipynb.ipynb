Step 3: Load Stable Diffusion Model

```python
pipe=StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
     torch_dtype=torch.float16,
    use_safetensors=True
)
pipe=pipe.to("cuda")
```

Step 5: Generate Images

Take user input:

Workflow Diagram
Input → User provides text prompt
Processing → Stable Diffusion generates image
Output → Display + Save image

```python
user_prompt= input("Enter Your image Prompt:")
```

Generate and display:

```python
image= pipe(user_prompt).images[0]
#Disply the generated image
plt.imshow(image)
plt.axis("off")
plt.title("Generated Image")
plt.show()
```
